```python
with open("initialdata.txt", encoding='utf-8') as f:
    data = f.read()
```

```python
data
```

```python
chunks = data.split("\n\n") 
chunks = [c for c in chunks if len(c)>3]
# chunks[0]
```

```python
def parse_chunk(chunk): 
        chunk = chunk.strip()
        sep_chunk = chunk.split('\n')
        username = sep_chunk[0]
        no_of_posts = int(sep_chunk[1].split("post")[0].replace(",",""))
        no_of_followers = int(sep_chunk[2].split("followers")[0].replace("K","000").replace("M","000000").replace(",","").replace(".",""))
        no_of_following = int(sep_chunk[3].split("following")[0].replace("K","000").replace("M","000000").replace(",","").replace(".",""))
        name = sep_chunk[4]
        if(len(sep_chunk)> 5):
            type_of_page = sep_chunk[5]
            bio = "\n".join(sep_chunk[6:])
        else:
            type_of_page = "Unknown"
            bio = ""
        # print(username, no_of_posts, no_of_followers, no_of_following, name, type_of_page, bio, sep="\n")
        return {"username": username, "no_of_posts": no_of_posts, "no_of_followers": no_of_followers, 
                "no_of_following": no_of_following, "name": name, "type_of_page": type_of_page, "bio": bio}
```

```python
print(parse_chunk(chunks[1]))
```

```python
all_chunks = [] 
for chunk in chunks:  
    parsed_chunk = parse_chunk(chunk)
    all_chunks.append(parsed_chunk)
print(all_chunks)
```

```python
# max posts
max=0
for chunck in all_chunks:
    if(max<chunck['no_of_posts']):
        max=chunck['no_of_posts']
        max_post_chunck=chunck
print(max_post_chunck)
```

```python
# max followers
max=0
for chunck in all_chunks:
    if(max<chunck['no_of_followers']):
        max=chunck['no_of_followers']
        max_followers_chunck=chunck
print(max_followers_chunck)
```

```python
# max following
max=0
for chunck in all_chunks:
    if(max<chunck['no_of_following']):
        max=chunck['no_of_following']
        max_following_chunck=chunck
print(max_following_chunck)
```
